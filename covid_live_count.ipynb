```python
import pandas as pd
import json, requests
import  matplotlib.pyplot as plt
covid_in_india_state_wise = requests.get('https://api.rootnet.in/covid19-in/unofficial/covid19india.org/statewise').json()
df_india_state_wise = pd.io.json.json_normalize(covid_in_india_state_wise['data']['statewise'])
df_india_state_wise.groupby('state').active.sum().plot(kind='bar', figsize=(20,4))
plt.show()

print("Total Active Cases ==> ",df_india_state_wise.active.sum())
print("Total Casualities  ==> ",df_india_state_wise.deaths.sum())
print("Total Recovered  ==> ",df_india_state_wise.recovered.sum())
```
